```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data (if not already loaded)
df = pd.read_csv('/content/sample_data/worldpopulationdata.csv')

# Select only necessary columns
df_2022 = df[['Country Name', '2022']].dropna()

# Sort to show Top 10 most populated countries
top10 = df_2022.sort_values(by='2022', ascending=False).head(10)

# Bar chart
plt.figure(figsize=(12,8))
sns.barplot(x='Country Name', y='2022', data=top10, palette='pastel')
plt.title('Population Distribution by Country (Top 10) - 2022')
plt.xticks(rotation=45)
plt.xlabel('Country Name')
plt.ylabel('Population in 2022')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/sample_data/worldpopulationdata.csv')

# Use top 10 most populated countries in 2022
df_2022 = df[['Country Name', '2022']].dropna()
top10 = df_2022.sort_values(by='2022', ascending=False).head(10)

# Bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='2022', y='Country Name', data=top10, palette='viridis')
plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Country Name')
plt.tight_layout()
plt.show()
```

```python
# Histogram of population values in 2022
plt.figure(figsize=(10, 6))
sns.histplot(df['2022'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Distribution of Country Populations in 2022')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
